```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
from IPython.display import display, Math
warnings.filterwarnings('ignore')
```

# Задание 1. Вещественное

```python
t = np.linspace(-10, 10, 1000)
w = np.linspace(-10, 10, 1000)
```

```python
def ft(w, func, a, b, c = None):
    if c != None:
        return np.array([1 / np.sqrt(2 * np.pi) * np.trapz(func(t, a, b, c) * np.exp(-1j * x * t), t) for x in w])
    return np.array([1 / np.sqrt(2 * np.pi) * np.trapz(func(t, a, b) * np.exp(-1j * x * t), t) for x in w])
```

## Прямоугольная функция

```python
def rectangular(t, a, b):
    return np.piecewise(t, [np.abs(t) <= b, np.abs(t) > b], [a, 0])
```

### Графики функции и её Фурье-образа

```python
fig = plt.figure(constrained_layout=True, figsize=(10, 10))

subfigs = fig.subfigures(2, 2)
c = [(1, 1), (1, 2), (2, 1), (2, 2)]
for outerind, subfig in enumerate(subfigs.flat):
    a, b = c[outerind]
    subfig.suptitle(f'$a$={a}, $b$={b}')
    axs = subfig.subplots(2, 1)
    plt.sca(axs[0])
    plt.plot(t, rectangular(t, a, b))
    plt.xlabel('$t$')
    plt.ylabel('$f(t)$')
    plt.grid(True)
    plt.sca(axs[1])
    plt.plot(w, ft(w, rectangular, a, b))
    plt.xlabel('$\\omega$')
    plt.ylabel('$\\hat f(\\omega)$')
    plt.grid(True)
```

### Равенство Парсеваля

```python
display(Math(f'||f||_2 - ||\mathcal{{F}}f||_2 = {np.round(np.sqrt(np.trapz(rectangular(t, 2, 1) * rectangular(t, 2, 1), t)) - np.sqrt(np.real(np.trapz(ft(w, rectangular, 2, 1) * np.conjugate(ft(w, rectangular, 2, 1)), w)))
, 4)}'))
```

## Треугольная функция

```python
def triangular(t, a, b):
    return np.piecewise(t, [np.abs(t) <= b, np.abs(t) > b], [lambda t: a - np.abs(a * t / b), 0])
```

### Графики функции и её Фурье-образа

```python
fig = plt.figure(constrained_layout=True, figsize=(10, 10))

subfigs = fig.subfigures(2, 2)
c = [(1, 1), (1, 2), (2, 1), (2, 2)]
for outerind, subfig in enumerate(subfigs.flat):
    a, b = c[outerind]
    subfig.suptitle(f'$a$={a}, $b$={b}')
    axs = subfig.subplots(2, 1)
    plt.sca(axs[0])
    plt.plot(t, triangular(t, a, b))
    plt.xlabel('$t$')
    plt.ylabel('$f(t)$')
    plt.grid(True)
    plt.sca(axs[1])
    plt.plot(w, ft(w, triangular, a, b))
    plt.xlabel('$\\omega$')
    plt.ylabel('$\\hat f(\\omega)$')
    plt.grid(True)
```

### Равенство Парсеваля

```python
display(Math(f'||f||_2 - ||\mathcal{{F}}f||_2 = {np.round(np.sqrt(np.trapz(triangular(t, 2, 1) * triangular(t, 2, 1), t)) - np.sqrt(np.real(np.trapz(ft(w, triangular, 2, 1) * np.conjugate(ft(w, triangular, 2, 1)), w)))
, 4)}'))
```

## Кардинальный синус

```python
def sinc(t, a, b):
    return a * np.sinc(b * t)
```

## Функция Гаусса

```python
def gaussian(t, a, b):
    return a * np.exp(-b * t ** 2)
```

# Задание 2. Комплексное

```python
a, b = 1, 1
```

```python
def g(t, a, b, c):
    return sinc(t + c, a, b)
```

## Графики функции и графики мнимой и вещественной части Фурье-образа

```python
fig = plt.figure(constrained_layout=True, figsize=(10, 10))

subfigs = fig.subfigures(2, 2)
C = [-2, -1, 0, 1]
for outerind, subfig in enumerate(subfigs.flat):
    c = C[outerind]
    subfig.suptitle(f'$c$={c}')
    axs = subfig.subplots(2, 1)
    plt.sca(axs[0])
    plt.plot(t, g(t, a, b, c))
    plt.xlabel('$t$')
    plt.ylabel('$g(t)$')
    plt.grid(True)
    plt.sca(axs[1])
    re = np.real(ft(w, g, a, b, c))
    im = np.imag(ft(w, g, a, b, c))
    plt.plot(w, re, label='$Re\\hat f(\\omega)$')
    plt.plot(w, im, label='$Im\\hat f(\\omega)$')
    plt.xlabel('$\\omega$')
    plt.legend()
    plt.grid(True)
```

## Графики модуля Фурье-образа

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 4))
fig.tight_layout(pad = 3)
for i, c in enumerate(C):
    plt.sca(axs[i])
    plt.title(f'$c$={c}')
    plt.plot(w, np.abs(ft(w, g, a, b, c)))
    plt.xlabel('$\\omega$')
    plt.ylabel('$|\\hat g(\\omega)|$')
    plt.grid(True)
```
